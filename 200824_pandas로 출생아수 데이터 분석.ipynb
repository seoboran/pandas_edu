## 1.1 Tidy Data 형식으로 만들기
- columns에 있는 값을 행으로 녹이는 것 

```python
df_kosis.head(1)
```

```python
df=df_kosis.melt(id_vars='시군구별')
df.head()
```

```python
df['시군구별'].unique()
```

```python
df[df['시군구별']=='시군구별']
```

```python
df[df['시군구별']!='시군구별']
```

```python
print(df.shape)
df=df[df['시군구별']!='시군구별'].copy()
df.shape
```

```python
df['시군구별'].unique()
```

```python
df.head()
```

```python
df['variable'].str.split(".", expand=True)
```

```python
df['연도']=df['variable'].str.split(".", expand=True)[0]
df['월']=df['variable'].str.split(".", expand=True)[1]
df['성별']=df['variable'].str.split(".", expand=True)[2]
df.head(3)
```

```python
df.tail()
```

```python
df['성별'].unique()
```

```python
df['성별'].nunique() # 1과 2 두개의 value가 들어있음
```

```python
df['성별']=df['성별'].fillna('전체')
df['성별'].unique()
```

```python
df['성별']=df['성별'].replace('1','남자').replace('2','여자')
df['성별'].unique() 
# 항상 다시 df['성별']에 저장을 해줘야 변경된 값이 반영이 된다.
#df['성별']= 꼭 해주기
```

```python
df['성별'].value_counts()
```

```python
df.head()
```

```python
df=df.rename(columns={'variable':'기간','value':'출생아수'})
df.head()
```

### 연도,월,성별별 출생아수 분석 

```python
df['출생아수'] #dbject type으로 되어있음 -> 정수로 바꿔줘야함 
```

```python
import numpy as np

# 결측치로 바꾼것을 다시 df['출생아수']에 적용
df['출생아수']=df['출생아수'].replace('-',np.nan)


#df['출생아수'].astype(int)
# - 로 인해서 정수로 변환이 안되있으므로 다시 정수로 변환 -> 에러 발생
```

```python
type(np.nan) #flaot 함수를 int로 바로 바꿔줄수없음
```

```python
df['출생아수']=df['출생아수'].replace('-',np.nan)
df['출생아수']=df['출생아수'].astype(float)
df['출생아수'].describe()
```

```python
df
```

```python
df['시군구별']=='전국'
```

```python
df[df['시군구별']=='전국']
```

```python
# 시군구별은 전국과 성별은 전체인 데이터만 가져오기 
df_all=df[(df['시군구별']=='전국')&(df['성별']=='전체')]
df_all.head(2)
```
